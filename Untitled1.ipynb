# RFM Customer Segmentation Analysis

This project segments ~4,338 customers from a UK-based online retailer using RFM (Recency, Frequency, Monetary) analysis. After cleaning ~25% of rows with missing CustomerIDs and removing returns/cancellations, customers were scored on a 1-5 scale and grouped into actionable segments. Champions (19% of customers) drive the highest average revenue (£6,713), while At Risk customers — despite high historical value — haven't purchased in over 4 months, representing a clear retention opportunity.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('online_retail.csv', encoding='ISO-8859-1')
df.head()
```

```python
print(df.shape)
df.info()
print(df.isnull().sum())
```

## Data Cleaning

```python
# Drop rows with missing CustomerID (can't do RFM without it)
df_clean = df.dropna(subset=['CustomerID'])

# Check for negative/zero quantities (returns or errors) and negative prices
print("Negative quantities:", (df_clean['Quantity'] <= 0).sum())
print("Negative/zero prices:", (df_clean['UnitPrice'] <= 0).sum())

print(df_clean.shape)
```

```python
df_clean = df_clean[(df_clean['Quantity'] > 0) & (df_clean['UnitPrice'] > 0)]

# Create a TotalPrice column (this is your "Monetary" building block)
df_clean['TotalPrice'] = df_clean['Quantity'] * df_clean['UnitPrice']

print(df_clean.shape)
df_clean.head()
```

## RFM Calculation

```python
import datetime as dt

# Convert InvoiceDate to actual datetime
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'])

# Reference date = one day after the last transaction in the dataset
reference_date = df_clean['InvoiceDate'].max() + dt.timedelta(days=1)
print("Reference date:", reference_date)

# Group by CustomerID to calculate R, F, M
rfm = df_clean.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,   # Recency
    'InvoiceNo': 'nunique',                                      # Frequency
    'TotalPrice': 'sum'                                          # Monetary
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.head(10)
```

## Scoring & Segmentation

```python
# Score each metric into quintiles (1-5)
# Recency: lower is better, so we reverse the labels
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])

# Combine into a single RFM score
rfm['RFM_Score'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

rfm.head(10)
```

```python
def segment_customer(row):
    if row['RFM_Score'] in ['555', '554', '544', '545', '454', '455', '445']:
        return 'Champions'
    elif row['R_score'] in [4,5] and row['F_score'] in [4,5]:
        return 'Loyal Customers'
    elif row['R_score'] in [4,5] and row['F_score'] in [1,2,3]:
        return 'Potential Loyalist'
    elif row['R_score'] == 5 and row['F_score'] == 1:
        return 'New Customers'
    elif row['R_score'] in [2,3] and row['F_score'] in [2,3]:
        return 'Needs Attention'
    elif row['R_score'] in [1,2] and row['F_score'] in [4,5]:
        return 'At Risk'
    elif row['R_score'] == 1 and row['F_score'] == 1:
        return 'Lost'
    else:
        return 'Others'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)

# See segment sizes
rfm['Segment'].value_counts()
```

## Visualizations

```python
import matplotlib.pyplot as plt

segment_counts = rfm['Segment'].value_counts()

plt.figure(figsize=(10,6))
segment_counts.plot(kind='bar', color='steelblue')
plt.title('Customer Segments (RFM Analysis)')
plt.xlabel('Segment')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
segment_summary = rfm.groupby('Segment').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean'
}).round(1).sort_values('Monetary', ascending=False)

segment_summary
```

```python
import seaborn as sns

# Pivot table: R_score vs F_score, values = average Monetary
heatmap_data = rfm.pivot_table(index='R_score', columns='F_score', values='Monetary', aggfunc='mean')

plt.figure(figsize=(9,6))
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlGnBu')
plt.title('Average Monetary Value by Recency & Frequency Score')
plt.xlabel('Frequency Score')
plt.ylabel('Recency Score')
plt.tight_layout()
plt.show()
```

```python
df_clean['InvoiceMonth'] = df_clean['InvoiceDate'].dt.to_period('M').astype(str)

monthly_revenue = df_clean.groupby('InvoiceMonth')['TotalPrice'].sum()

plt.figure(figsize=(11,5))
monthly_revenue.plot(kind='line', marker='o', color='darkorange')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue (£)')
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=rfm, x='Frequency', y='Monetary', hue='Segment', palette='tab10', alpha=0.7)
plt.title('Frequency vs Monetary by Segment')
plt.xlabel('Frequency (number of orders)')
plt.ylabel('Monetary (£)')
plt.tight_layout()
plt.show()
```
